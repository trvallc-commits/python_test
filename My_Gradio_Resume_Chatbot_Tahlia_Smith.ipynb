```python
import gradio as gr


# DEFINE FUNCTIONS (Cell 6)

def process_uploaded_pdf(file_obj):
    """Extracts text from the uploaded PDF, chunks it, and builds a search index."""
    global query_engine  # Allows the chatbot function below to access this index

    if file_obj is None:
        return "⚠️ No file uploaded."

    try:
        # Check if file_obj is a Gradio file dictionary or a raw object
        if isinstance(file_obj, dict):
            file_path = file_obj.get("path") or file_obj.get("name")
        else:
            file_path = getattr(file_obj, "name", None) or getattr(file_obj, "path", None)

        if not file_path:
            return "❌ Could not resolve the temporary file path."

        # Load and read the uploaded PDF file using the resolved path
        reader = PdfReader(file_path)
        pages = [page.extract_text() for page in reader.pages]
        doc_pages = [{"page_num": i, "text": p} for i, p in enumerate(pages)]

        # Use the working sliding window approach to chunk the text
        full_contract_text = "\n\n".join([doc["text"] for doc in doc_pages])
        splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=100)
        chunks = splitter.split_text(full_contract_text)

        # Structure the data using LlamaIndex nodes
        nodes = []
        for chunk in chunks:
            node = TextNode(text=chunk, metadata={"doc_type": "Contract"})
            nodes.append(node)

        # Rebuild the vector store index using the explicitly defined Google embedding model
        new_index = VectorStoreIndex(nodes, embed_model=current_embed_model)

        # Create query engine, to pass the current_llm
        query_engine = new_index.as_query_engine(llm=current_llm)

        return f"✅ Success! Split into {len(chunks)} chunks. Ask your questions!"
    except Exception as e:
        return f"❌ Error processing PDF: {e}"

def chat_with_contract(user_message, chat_history):
    """Sends the user message to the active query engine and logs chat history."""
    if not user_message.strip():
        return "", chat_history
    try:
        # Route question to Gemini 2.5 Flash through LlamaIndex query engine
        response = query_engine.query(user_message)
        bot_response = response.response
    except Exception as e:
        bot_response = f"Please upload and process a PDF first. Error: {e}"

    # Modern Gradio Chatbots use a dictionary structure with 'role' & 'content' keys
    chat_history.append({"role": "user", "content": user_message})
    chat_history.append({"role": "assistant", "content": bot_response})

    # Return values in order: clear the textbox, update the history format
    return "", chat_history



# BUILD THE VISUAL LAYOUT


with gr.Blocks(theme=custom_theme) as demo:
    gr.Markdown("# 📜 My Resume AI Assistant")
    gr.Markdown("I am an AI agent trained on my professional background, projects, and skills. Feel free to type a question below or click one of the example prompts to get started!")

    # Create a horizontal row to split the screen left and right
    with gr.Row():

        # LEFT COLUMN: Narrower side for file uploading and processing
        with gr.Column(scale=1):
            pdf_input = gr.File(label="📄 Upload PDF", file_types=[".pdf"])
            process_btn = gr.Button("🔄 Process Document")
            upload_status = gr.Textbox(label="Status", placeholder="Awaiting file processing...")

        # RIGHT COLUMN: Wider side for your chat conversation view
        with gr.Column(scale=2):
            chatbot = gr.Chatbot(label="Chat History", height=400)
            user_input = gr.Textbox(placeholder="Ask a question about your document...", label="Your Question")
            with gr.Row(): # New row for buttons
                send_btn = gr.Button("📤 Send", variant="primary")
                clear_input_btn = gr.Button("Clear Input") # New button

            # Add gr.Examples for the user_input textbox
            gr.Examples(
                examples=[
                    "What is your technical skill set?",
                    "Can you tell me about your most recent project?",
                    "What is your educational background?",
                    "Why would you be a great fit for a software engineering role?"
                ],
                inputs=user_input,
                label="Try these example questions:"
            )


#  CONNECT THE BUTTONS TO THE FUNCTIONS


    # Connect "Process Document" button to processing function
    process_btn.click(       # to process document or pdf
        fn=process_uploaded_pdf,
        inputs=pdf_input,
        outputs=upload_status
    )

    # Connect your Chat box submission events
    send_btn.click(fn=chat_with_contract, inputs=[user_input, chatbot], outputs=[user_input, chatbot])
    user_input.submit(fn=chat_with_contract, inputs=[user_input, chatbot], outputs=[user_input, chatbot])
    clear_input_btn.click(lambda: "", outputs=user_input) # Connect clear button


# CRITICAL RESET & LAUNCH SEQUENCE


# Force-close any old stuck ports before opening a new one
gr.close_all()

# Launch the final complete web dashboard interface
demo.launch(share=True, debug=True, js=force_light_mode, theme=custom_theme)
```
